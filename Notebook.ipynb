## Questions

Now let's write and execute SQL queries to solve Questions

### Question 1

##### Find the total number of crimes recorded in the CRIME table.


```python
cur.execute('SELECT COUNT(ID) FROM CHICAGO_CRIME_DATA')
cur.fetchone()
```

### Question 2

##### List community areas with per capita income less than 11000.


```python
cur.execute('''SELECT COMMUNITY_AREA_NAME FROM CENSUS_DATA
                WHERE PER_CAPITA_INCOME < 11000''')
#Convert result of Query to Pandas Dataframe and rename its Column
pd.DataFrame(cur.fetchall()).rename(columns={0:'COMMUNITY_AREA_NAME'})
```

### Question 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)


```python
# Help to find right value
CHICAGO_CRIME_DATA['PRIMARY_TYPE'].unique()
```

```python
cur.execute('''SELECT CASE_NUMBER FROM CHICAGO_CRIME_DATA 
                WHERE PRIMARY_TYPE= 'OFFENSE INVOLVING CHILDREN' ''')
pd.DataFrame(cur.fetchall()).rename(columns={0:'CASE_NUMBER'})
```

### Question 4

##### List all kidnapping crimes involving a child?


```python
cur.execute('''SELECT * FROM CHICAGO_CRIME_DATA 
                WHERE PRIMARY_TYPE= 'KIDNAPPING' AND DESCRIPTION LIKE '%CHILD%' ''')
# Retrieve all columns name and set them to df 
field_names = [i[0] for i in cur.description]
df = pd.DataFrame(cur.fetchall())
df.columns=[field_names]
df
```

### Question 5

##### What kinds of crimes were recorded at schools?


```python
cur.execute('''SELECT DISTINCT PRIMARY_TYPE FROM CHICAGO_CRIME_DATA 
                WHERE LOCATION_DESCRIPTION LIKE '%SCHOOL%' ''')
pd.DataFrame(cur.fetchall()).rename(columns={0:'PRIMARY_TYPE'})
```

### Question 6

##### List the average safety score for each type of school.


```python
CHICAGO_PUBLIC_SCHOOLS['Elementary, Middle, or High School'].unique()
```

```python
cur.execute('''SELECT "Elementary, Middle, or High School",
                avg(SAFETY_SCORE) FROM CHICAGO_PUBLIC_SCHOOLS
                GROUP BY "Elementary, Middle, or High School" ''')
pd.DataFrame(cur.fetchall()).rename(columns={0:'Type of School',1:'Average Safety'})
```

### Question 7

##### List 5 community areas with highest % of households below poverty line


```python
cur.execute('''SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY
                FROM CENSUS_DATA
                ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY  DESC LIMIT 5 ''')
pd.DataFrame(cur.fetchall()).rename(columns={0:'COMMUNITY_AREA_NAME',1:'PERCENT_HOUSEHOLDS_BELOW_POVERTY'})
```

### Question 8

##### Which community area is most crime prone?


```python
cur.execute(''' SELECT C ,NUM FROM (SELECT COUNT(*) AS C,
                COMMUNITY_AREA_NUMBER AS NUM FROM CHICAGO_CRIME_DATA
                GROUP BY COMMUNITY_AREA_NUMBER) ORDER BY C DESC LIMIT 2 ''')
pd.DataFrame(cur.fetchall()).rename(columns={0:'Count Of Crimes',1:'COMMUNITY_AREA_NUMBER'})
# WE HAVE 2 MAX DATA, ONE WITH NAN AREA NUMBER AND OTHER 25, SO I RETRIEVE AREA_NAME
# OF SECOND ONE( I MEAN 25 ) IN THE QUERY BELOW
```

```python
cur.execute(''' SELECT COMMUNITY_AREA_NAME, C FROM
                (SELECT C ,NUM FROM (SELECT COUNT(*) AS C,
                COMMUNITY_AREA_NUMBER AS NUM FROM CHICAGO_CRIME_DATA
                GROUP BY COMMUNITY_AREA_NUMBER) ORDER BY C DESC LIMIT 2) A
                LEFT JOIN CENSUS_DATA B WHERE A.NUM = B.COMMUNITY_AREA_NUMBER ''')

pd.DataFrame(cur.fetchall()).rename(columns={0:'COMMUNITY_AREA_NAME',1:'Count Of Crimes'})
```

### Question 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
cur.execute(''' SELECT COMMUNITY_AREA_NAME, MAX FROM
                (SELECT COMMUNITY_AREA_NAME, MAX(HARDSHIP_INDEX) AS MAX 
                    FROM CENSUS_DATA) ''')

pd.DataFrame(cur.fetchall()).rename(columns={0:'COMMUNITY_AREA_NAME',1:'MAX_HARDSHIP_INDEX'})
```

### Question 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
cur.execute(''' SELECT COMMUNITY_AREA_NAME, C FROM
                (SELECT C ,NUM FROM (SELECT COUNT(*) AS C,
                COMMUNITY_AREA_NUMBER AS NUM FROM CHICAGO_CRIME_DATA
                GROUP BY COMMUNITY_AREA_NUMBER) ORDER BY C DESC LIMIT 2) A
                LEFT JOIN CENSUS_DATA B WHERE A.NUM = B.COMMUNITY_AREA_NUMBER ''')
pd.DataFrame(cur.fetchall()).rename(columns={0:'COMMUNITY_AREA_NAME',1:'Count Of Crimes'})
```

### Question 11
Write and execute a SQL query to list the school names, community names and average attendance for communities with a hardship index of 98.

```python
cur.execute('''SELECT B.NAME_OF_SCHOOL, A.COMMUNITY_AREA_NAME ,B.AVERAGE_STUDENT_ATTENDANCE
                FROM CENSUS_DATA A LEFT JOIN CHICAGO_PUBLIC_SCHOOLS B
                ON A.COMMUNITY_AREA_NUMBER = B.COMMUNITY_AREA_NUMBER
                WHERE A.HARDSHIP_INDEX = 98''')

pd.DataFrame(cur.fetchall()).rename(columns={0:'NAME_OF_SCHOOL', 1:'COMMUNITY_AREA_NAME'
                                          ,2:'AVERAGE_STUDENT_ATTENDANCE'})
```

### Question 12
Write and execute a SQL query to list all crimes that took place at a school. Include case number, crime type and community name.

```python
cur.execute('''SELECT C.CASE_NUMBER, C.PRIMARY_TYPE ,A.COMMUNITY_AREA_NAME, C.LOCATION_DESCRIPTION
                FROM CHICAGO_CRIME_DATA C LEFT JOIN CENSUS_DATA A
                ON C.COMMUNITY_AREA_NUMBER = A.COMMUNITY_AREA_NUMBER
                WHERE C.LOCATION_DESCRIPTION LIKE '%SCHOOL%' ''')

pd.DataFrame(cur.fetchall()).rename(columns={0:'CASE_NUMBER',1:'PRIMARY_TYPE',
                                             2:'COMMUNITY_AREA_NAME',3:'LOCATION_DESCRIPTION'})
```

# Creating a View

### Question 1
Write and execute a SQL statement to create a view showing the columns listed in the following table, with new column names as shown in the second column.

Column name in CHICAGO_PUBLIC_SCHOOLS ----> Column name in view     
NAME_OF_SCHOOL ----> School_Name     
Safety_Icon ----> Safety_Rating     
Family_Involvement_Icon ----> Family_Rating     
Environment_Icon ----> Environment_Rating     
Instruction_Icon ----> Instruction_Rating          
Leaders_Icon ----> Leaders_Rating     
Teachers_Icon ----> Teachers_Rating     

-Write and execute a SQL statement that returns all of the columns from the view.     
-Write and execute a SQL statement that returns just the school name and leaders rating from the view.

```python
cur.execute('DROP VIEW  IF EXISTs view1' )

cur.execute('''CREATE VIEW view1 (School_Name,Safety_Rating,Family_Rating, Environment_Rating 
                                   ,Instruction_Rating ,Leaders_Rating ,Teachers_Rating)
                AS SELECT NAME_OF_SCHOOL, Safety_Icon, Family_Involvement_Icon, Environment_Icon
                           ,Instruction_Icon,Leaders_Icon, Teachers_Icon 
                FROM CHICAGO_PUBLIC_SCHOOLS''')

cur.execute('SELECT * FROM view1')

# get col names from view and set to dataframe
field_names = [i[0] for i in cur.description]
df = pd.DataFrame(cur.fetchall())
df.columns=[field_names]
print(df.shape)
```
